## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
data.drop(['Customer_ID', 'Name', 'SSN', 'Month', 'Payment_of_Min_Amount'],axis= 1 ,inplace=True) #Dropping columns irrelevant for analysis
#Using IQR method, removing outliers from the annual income column
Q1 = data['Annual_Income'].quantile(0.25)
Q3 = data['Annual_Income'].quantile(0.75)
IQR = Q3-Q1
data = data[(data['Annual_Income']>=(Q1-1.5*IQR))&(data['Annual_Income']<=(Q3+1.5*IQR))]
data['Debt_to_Income_ratio']=data['Outstanding_Debt']/data['Annual_Income']# Introduction of new column for analysis - Debt to Income Ratio

def classify_income_bracket(income): #Custom function to create income segmentation
  if income <= 20000:
    return "Low"
  elif income<=50000:
    return "Medium"
  elif income<=100000:
    return "High"
  else:
    return "Very High"

data['Income_Bracket'] = data['Annual_Income'].apply(classify_income_bracket)  #introduction of new column for analysis - income bracket
new_order = ['Low','Medium','High','Very High']
data['Income_Bracket'] = pd.Categorical(data['Income_Bracket'], categories = new_order, ordered = True) #Ensuring order of the income bracket is correct to enable accurate representation of the same on labels in visualisation charts.
````

### What all manipulations have you done and insights you found?

The following manipulation has been carried out on the dataset :

1. Dropping irrelevant columns : Customer_ID, Name, SSN, Month and Payment of minimum amount are primarily customer related attributes unrelated to attributes which focus on financial or behavorial features.

2. Removal of outliers from Annual Income column using Inter Quartile Range (IQR) Method to ensure analysis is not skewed by existence of outliers in the dataset

3. Introduction of a new column - Debt to Income Ratio - an important financial ratio for assessing liquidity and creditworthiness of customer.

4. Introduction of a new column - Income Bracket - to faciliate income group wise targeted analysis.

**Insights:**

1.Outlier Management: Removing outliers allows for a more accurate representation of customer income, leading to more reliable insights.

2.Debt-to-Income Ratio: This new metric provides a clearer understanding of customers' financial health, indicating potential credit risks.

3.Income Segmentation: The creation of income brackets enables targeted analysis, making it easier to tailor financial product recommendations based on income levels.
